### Solve Ax = b
```
Parameters:
A: list of lists
    Coefficient matrix (n x n)
b: list
    Right-hand side vector (n)
x0: list
    Initial guess (default: zeros)
tol: float
    Convergence tolerance
max_iter : int
    Maximum number of iterations

Returns:
x: list
Approximate solution vector
```

````python
import numpy as np
````

### 1. Inisialisasi Fungsi `Eliminasi Gauss`, `Gauss-Jordan`, `Gauss-Seidel`, dan `Iterasi Jacob`

### a. Fungsi Eliminasi Gauss

````python
def gauss_elimination(a, b):
    n = len(b)

    # Forward elimination
    for i in range(n):
        # Pivoting
        if a[i][i] == 0:
            for j in range(i + 1, n):
                if a[j][i] != 0:
                    a[i], a[j] = a[j], a[i]
                    b[i], b[j] = b[j], b[i]
                    break

        pivot = a[i][i]
        if pivot == 0:
            raise ValueError("Matrix is singular or nearly singular.")

        # Normalize pivot row
        for j in range(i, n):
            a[i][j] /= pivot
        b[i] /= pivot

        # Eliminate below
        for k in range(i + 1, n):
            factor = a[k][i]
            for j in range(i, n):
                a[k][j] -= factor * a[i][j]
            b[k] -= factor * b[i]

    # Back substitution
    x = [0 for _ in range(n)]
    for i in range(n - 1, -1, -1):
        x[i] = b[i]
        for j in range(i + 1, n):
            x[i] -= a[i][j] * x[j]

    return x
````

### b. Fungsi Gauss Jordan

````python
def gauss_jordan(a, b):
    n = len(b)
    a_copy = copy.deepcopy(a)

    # Augment matrix A with vector b
    for i in range(n):
        a_copy[i].append(b[i])

    # Perform Gauss-Jordan elimination
    for i in range(n):
        # Pivoting
        if a_copy[i][i] == 0:
            for j in range(i + 1, n):
                if a_copy[j][i] != 0:
                    a_copy[i], a_copy[j] = a_copy[j], a_copy[i]
                    break

        pivot = a_copy[i][i]
        if pivot == 0:
            raise ValueError("Matrix is singular or nearly singular.")

        # Normalize pivot row
        for j in range(i, n + 1):
            a_copy[i][j] /= pivot

        # Eliminate all other rows
        for k in range(n):
            if k != i:
                factor = a_copy[k][i]
                for j in range(i, n + 1):
                    a_copy[k][j] -= factor * a_copy[i][j]

    # Extract solution
    x = [a_copy[i][n] for i in range(n)]
    return x
````

### c. Fungsi Gauss Seidel

````python
def gauss_seidel(A, b, x0=None, tol=1e-6, max_iter=100):
    n = len(b)
    # initial guess
    x = [0.0] * n if x0 is None else x0[:]

    for k in range(max_iter):
        x_new = x[:]
        for i in range(n):
            sum1 = sum(A[i][j] * x_new[j] for j in range(i))  # updated values
            sum2 = sum(A[i][j] * x[j] for j in range(i + 1, n))  # old values
            x_new[i] = (b[i] - sum1 - sum2) / A[i][i]

        # convergence check (infinity norm)
        if max(abs(x_new[i] - x[i]) for i in range(n)) < tol:
            print(f"Converged in {k + 1} iterations")
            return x_new
        x = x_new

    print("Did not converge within max_iter")
    return x
````

### d. Fungsi Iterasi Jacobi

````python
def jacobi(A, b, x0=None, tol=1e-10, max_iter=100):
    n = len(b)
    # initial guess
    x = [0.0] * n if x0 is None else x0[:]

    for k in range(max_iter):
        x_new = [0.0] * n
        for i in range(n):
            # compute sum excluding diagonal element
            sum_terms = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - sum_terms) / A[i][i]

        # convergence check (infinity norm)
        if max(abs(x_new[i] - x[i]) for i in range(n)) < tol:
            print(f"Converged in {k + 1} iterations")
            return x_new
        x = x_new

    print("Did not converge within max_iter")
    return x
````
